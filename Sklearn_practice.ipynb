### Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
import seaborn as sns
sns.set()
```

### Load my csv file

#### Explore the data

```python
raw_data = pd.read_csv('./energydata_complete.csv')
raw_data.describe(include='all')
```

### Drop the columns that are not 

```python
data = raw_data.drop(['date', 'lights'], axis=1)
```

```python
data.head()
```

```python
data.describe(include='all')
```

### Standardize the data and Create Linear regression model
#### Declare the input and targets
#### Scale the data

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
normalised_data = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)
features_data = normalised_data.drop(columns=['Appliances'], axis=1)
appliances_target = normalised_data['Appliances']
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(features_data,appliances_target, test_size = 0.3, random_state = 42)
reg = LinearRegression()
reg.fit(x_train,y_train)
predicted_values = reg.predict(x_test)
```

```python
predicted_values
```

### Mean Absolute Erroe

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 3)
```

### R-Squared

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 3)
```

### Residual Sum of Squares

```python
rss = np.sum(np.square(y_test - predicted_values))
round(rss, 3)
```

### Root Mean Squared Error

```python
from sklearn.metrics import  mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```
